## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=['EIN','NAME'])
application_df.head()
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
for_binning = application_df['APPLICATION_TYPE'].value_counts()
for_binning
```

```python
binning_df = pd.DataFrame(for_binning)
binning_df
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`

# Create list
application_types_to_replace = []

# Create bins
bins = [0, 200, 30000]

# Create bin names
group_names = ["being_replaced", "not_being_replaced"]

# Sort data into bins
binning_df['replacement_status'] = pd.cut(binning_df['count'], bins, labels=group_names)
binning_df
```

```python
binning_df.index[8]
```

```python
binning_df['replacement_status'][7]
```

```python
# Append the APPLICATION_TYPE values with group names of "being replaced" to the "application_types_to_replace" list
iter = 0
for replacing_value in binning_df['replacement_status']:
    if replacing_value == "being_replaced":
        application_types_to_replace.append(binning_df.index[iter])
        iter = iter + 1
    else:
        iter = iter + 1

application_types_to_replace
```

```python
# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
application_df.head()
```

```python
# Look at CLASSIFICATION value counts for binning
class_bins = application_df['CLASSIFICATION'].value_counts()
class_bins
```

```python
class_binning_df = pd.DataFrame(class_bins)
class_binning_df
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `classificatoins_to_replace`

# Create list
classifications_to_replace = []

# Create bins
bins = [0, 800, 20000]

# Create bin names
group_names = ["being_replaced", "not_being_replaced"]

# Sort data into bins
class_binning_df['replacement_status'] = pd.cut(class_binning_df['count'], bins, labels=group_names)
class_binning_df
```

```python
# Append the CLASSIFICATION values with group names of "being replaced" to the "classifications_to_replace" list
iter = 0
for replacing_value in class_binning_df['replacement_status']:
    if replacing_value == "being_replaced":
        classifications_to_replace.append(class_binning_df.index[iter])
        iter = iter + 1
    else:
        iter = iter + 1

classifications_to_replace
```

```python

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df = pd.get_dummies(application_df)
```

```python
# Define the features (X) and target (y) variables
X = application_df.drop(columns=['IS_SUCCESSFUL'])
y = application_df['IS_SUCCESSFUL']
```

```python
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net
number_input_features = len(X_train_scaled[0])
hidden_nodes_layer1 =  80
hidden_nodes_layer2 = 30
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE
nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu"))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
callback = tf.keras.callbacks.ModelCheckpoint(
    filepath='AlphabetSoupCharity_Optimization.keras', # Change filepath to end with .keras
    verbose=1,
    save_weights_only=False, # Change to False to save the entire model
    save_freq=5)
fit_model = nn.fit(X_train_scaled, y_train, epochs=100, callbacks=[callback])
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn.save("AlphabetSoupCharity.h5")
```
